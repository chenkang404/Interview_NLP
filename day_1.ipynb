```python
def isValid(s) -> bool:
    if len(s) == 0:
        return True
    stack = []
    for c in s:
        if c == "(" or c == "[" or c == "{":
            stack.append(c)
        else:
            if len(stack) == 0:
                return False
            else:
                temp = stack.pop()

                if c == ")":
                     if temp != "(":
                         return False
                if c == "]":
                    if temp != "[":
                        return False
                if c == "}":
                    if temp != "{":
                        return False
    return True if len(stack) == 0 else False

strings_ = "({}{}())"
isValid(strings_)
            
```

```python
#合并两个有序列表
class ListNode:
    def __init__(self,x):
        self.val = x
        self.next = None

def mergeTwoLists(l1:ListNode,l2:ListNode) -> ListNode:
    res = ListNode()
    cur = res 
    while(l1 != None and l2 != None):
        if l1.val <= l2.val:
            cur.next = l1
            l1 = l1.next
        else:
            cur.next = l2
            l2 = l2.next
        cur = cur.next
    cur.next = l1 or l2
    return res.next

def mergeTwoLists_v2(l1:ListNode,l2:ListNode) -> ListNode:
    if l1 == None or l2 == None:
        return l1 or l2
    if l1.val <= l2.val:
        l1.next = mergeTwoLists_v2(l1.next,l2)
        return l1

    else:
        l2 = mergeTwoLists_v2(l1,l2.next)
        return l2


def mergetwolist(l1:ListNode,l2:ListNode):
    if l1 == None or l2 == None:
        return l1 or l2
    if l1.val >=l2.val:
        l1.next = mergetwolist(l1.next,l2)
        return l1
    else:
        l2.next = mergetwolist(l1,l2.next)
        return l2
    
def mergetwolist_(l1:ListNode,l2:ListNode):
    res = ListNode()
    car = res

    while (l1 != None and l2 != None):
        if l1.val <= l2.val:
            res.next = l1
            l1 = l1.next
        else:
            res.next = l2
            l2 = l2.next
        car = car.next
    car.next = l1 or l2
    return res.next
```

```python
#p5

#backtracking
from typing import List


class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        """
        题目描述：给定 n 代表生成括号的对数，请你写出一个函数以生成所有可能的组合。
        例如，n = 3，生成结果为：
                        [
                        "((()))",
                        "(()())",
                        "(())()",
                        "()(())",
                        "()()()"
                        ]
        """

        res = []
        cur_str = ''

        def dfs(cur_str, left, right):
            """
            :param cur_str: 从根结点到叶子结点的路径字符串
            :param left: 左括号还可以使用的个数
            :param right: 右括号还可以使用的个数
            :return:
            """
            if left == 0 and right == 0:
                res.append(cur_str)
                return
            if right < left:
                return
            if left > 0:
                dfs(cur_str + '(', left - 1, right)
            if right > 0:
                dfs(cur_str + ')', left, right - 1)

        dfs(cur_str, n, n)
        return res
    
solution = Solution()
solution.generateParenthesis(3)
def gp_dfs(n):
    res = []
    cur_str = ""
    def dfs(cur_str,left,right):
        if left == 0 and right == 0:
            res.append(cur_str)
            return
        if right <left:
            return
        if left > 0:
            dfs(cur_str+"(",left-1,right)
        if right > 0:
            dfs(cur_str + ")",left,right-1)

    dfs(cur_str,n,n)
    return res
gp_dfs(3)
```

```python
#p6
#iteration/recursion

def swapPairs_v1(head: ListNode) -> ListNode:
    """time:O(N)
       space:O(1)
       eg:1->2->3->4,
       交换相邻节点,eg1->2->3->4,交换后2->1->4->3
    Args:
        head (ListNode): _description_

    Returns:
        ListNode: _description_
    """
    if head == None or head.next == None:
        return head
    res = ListNode()
    res.next = head #res->1->2->3->4
    cur = res 
    while cur.next !=None and cur.next.next != None:
        #step1:明确指针，cur=res,head=1,nxt=2,tmp=3
        nxt = head.next
        tmp = nxt.next
        #交换=>res->2->1->3->4
        cur.next = nxt
        nxt.next = head
        head.next = tmp
        #进行下次循环交换，即重新明确指针，cur=1,head=3
        cur = head
        head = head.next
    return res.next
    

def swapPairs(head: ListNode) -> ListNode:
    #Recursion
    if head == None or head.next == None:
        return head
    nxt = head.next
    head.next = swapPairs(head.next.next)
    nxt.next = head
    return nxt
```

```python
#P8
def searchInsert(nums, target):
    """time:O(N)搜索插入

    Args:
        nums (_type_): _description_
        target (_type_): _description_

    Returns:
        _type_: _description_
    """
    if nums == None or len(nums) == 0:
        return 0
    for i, num in enumerate(nums):
        if num>target:
            return i
    return len(nums)

def searchInsert_v2(nums, target):
    """time:O(logN),折半插入
    """
    if nums == None or len(nums) == 0:
        return 0
    left = 0
    right = len(nums) - 1
    while(left < right):
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            right = mid
        else:
            left = mid + 1
    return left if nums[left] >= target else left+1 
```

```python
#p9
def groupAnagrams(strs: List[str]) -> List[List[str]]:
    """N is the size of strs
       K is the size of each element in strs
       输入：["eat", "tea", "tan", "ate", "nat", "bat"]
        输出：[["ate", "eat", "tea"], ["nat", "tan"], ["bat"]]
       time: O(NKlogK)
    Args:
        strs (List[str]): _description_

    Returns:
        List[List[str]]: _description_
    """
    if len(strs) < 2:
        return [strs]
    result = {}
    for s in strs:
        temp = "".join(sorted(s))
        result[temp] = result.get(temp,[]) + [s]
    return result.values()

def groupAnagrams_v2(strs: List[str]) -> List[List[str]]:
    """
    ord是一个内置函数，用于获得单个字符的整数表示
    """
    if len(strs) < 2:
        return [strs]
    result = {}
    for s in strs:
        count_table = [0]*26
        for c in s:
            count_table[ord(c) - ord('a')] += 1
        key = tuple(count_table)
        result[key] = result.get(key,[]) + [s]
    return result
```

```python
class Solution:
    """
    斐波那契
    """
    def fib(self, n: int) -> int:
        if n==0 or n==1:
            return n
        a,b = 0,1
        for _ in range(n):
            a,b = b,a+b
        return a%1000000007
    
solution = Solution()
solution.fib(10)
```

```python
class ListNode:
    def __init__(self,x):
        self.val = x
        self.next = None

    
class Solution:
    """
    删除列表倒数第n个结点
    """
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        dummy = ListNode(0, head)
        left, right = dummy, head
        n -= 1
        while n: # 快慢指针，快指针先走
            n -= 1
            right = right.next
       
        while right.next: # 慢指针跟上
            right = right.next
            left = left.next
       
        left.next = left.next.next # 删除节点
        return dummy.next
```

```python
# 归并排序
def merge_sort(arr):
    """
    归并排序是一种分治算法，它将数组分成两半，对每一半进行排序，然后将排序好的两半合并。
    """
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# 示例
arr = [12, 11, 13, 5, 6, 7]
merge_sort(arr)
print("Sorted array is:", arr)
```

```python
def quick_sort(arr):
    """
    时间复杂度：O(nlogn)
    快速排序也是一种分治算法，通过选择一个“基准”元素，将数组分为两个子数组，
    一个包含小于基准的元素，另一个包含大于基准的元素。
    递归
    """
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr.pop()

    items_lower = [x for x in arr if x <= pivot]
    items_greater = [x for x in arr if x > pivot]

    return quick_sort(items_lower) + [pivot] + quick_sort(items_greater)

def quick_sort_(arr:List[int]):

    if len(arr) <= 1:
        return arr
    else:
        pivot = arr.pop()

    items_low = [x for x in arr if x<=pivot]
    items_hight = [y for y in arr if y>pivot]

    return quick_sort_(items_low) + [pivot] + quick_sort_(items_hight)

# 示例
arr = [10, 7, 8, 9, 1, 5]
# sorted_arr = quick_sort(arr)
# print("Sorted array is:", sorted_arr)
sorted_arr_2 = quick_sort_(arr)
print(sorted_arr_2)
print(arr)
```

```python
def insertion_sort(arr):
    #插入排序，每一次只保证前i个是有序的(arr[:i]有序)，对于i+1的值，会在arr[:i]中找到相应的位置进行插入
    #时间复杂度：o(n2)
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:#为将要插入的值，腾位置
            arr[j + 1] = arr[j]
            print(f"j:{j},key:{key},arr:{arr}")
            j -= 1
        # print(j)
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6]
print("Original array:", arr)
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
def bubble_sort(arr):
    """
    时间复杂度:O(n2)
    """
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        print(arr)

# Example usage
arr = [64, 34, 25, 12, 22, 11, 90]
print("Original array:", arr)
bubble_sort(arr)
print("Sorted array:", arr)
```

```python

def binary_search_first(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            result = mid
            right = mid - 1  # 继续在左侧查找更小的索引
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result

def binary_search_last(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            result = mid
            left = mid + 1  # 继续在右侧查找更大的索引
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result

# 示例
arr = [1, 2, 3, 4, 4, 4, 5, 6]
target = 4
print("First occurrence of", target, "is at index:", binary_search_first(arr, target))
print("Last occurrence of", target, "is at index:", binary_search_last(arr, target))
```

```python
def edit_distance(A, B):
    """
    "↑":删除操作
    "←":插入操作
    "↖":替换或不操作
    """
    m, n = len(A), len(B)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    op = [[""] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m + 1):
        dp[i][0] = i
        if i > 0:
            op[i][0] = "↑"

    for j in range(n + 1):
        dp[0][j] = j
        if j > 0:
            op[0][j] = "←"

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if A[i-1] == B[j-1]:
                dp[i][j] = dp[i-1][j-1]
                op[i][j] = "↖"
            else:
                del_op = dp[i-1][j] + 1
                ins_op = dp[i][j-1] + 1
                rep_op = dp[i-1][j-1] + 1

                dp[i][j] = min(del_op, ins_op, rep_op)
                
                if dp[i][j] == del_op:
                    op[i][j] = "↑"
                elif dp[i][j] == ins_op:
                    op[i][j] = "←"
                else:
                    op[i][j] = "↖"

    # Print the dp table and the operations
    print("DP Table:")
    for row in dp:
        print(row)
    
    print("\nOperation Table:")
    for row in op:
        print(row)

    return dp[m][n], op

# Example usage
A = "wt"
B = "aefwe"
distance, operations = edit_distance(A, B)
print(f"Edit Distance: {distance}")
```

```python
def knapsack(weights, values, capacity):
    """
    weights: 各物品的重量
    values: 各物品的值
    capacity: 重量限制
    """
    #背包问题，动态规划
    n = len(values)
    # 创建一个二维数组，行数为物品数量加1，列数为容量加1
    dp = [[0 for x in range(capacity + 1)] for x in range(n + 1)]

    # 填充 dp 数组
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                # w是当前重量限制,
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
            else:
                # 如果当前物品不能放入背包，继承上一个物品的结果
                dp[i][w] = dp[i-1][w]
    for line in dp:
        print(line)

    # 返回最大价值
    return dp[n][capacity]

# 示例
values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50
print("最大价值:", knapsack(weights, values, capacity))
```

```python
def knapsack_optimized(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)

    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):
            dp[w] = max(dp[w], dp[w - weights[i]] + values[i])

    return dp[capacity]

# Example usage
weights = [10, 20, 30]
values = [60, 100, 120]
capacity = 50
max_value = knapsack_optimized(weights, values, capacity)
print(f"The maximum value that can be put in a knapsack of capacity {capacity} is {max_value}.")
```

```python
#最长回文子串


def longset_pal(s):
    n = len(s)
    if n<2:
        return s
    
    dp = [[False]*n for _ in range(n)]
    start,max_len = 0,1
    for i in range(n):
        dp[i][i] = True

    for length in range(2,n+1):
        for i in range(n):
            j = i + length -1
            if j>n:
                break
            if s[j] == s[i]:
                #当 s[i] == s[j] 时，判断 s[i:j+1] 是否是回文需要看 s[i+1:j] 是否是回文。
                #dp[i+1][j-1] 表示子字符串 s[i+1:j] 是否是回文。
                if length == 2:
                    dp[i][j] = True
                else:
                    dp[i][j] = dp[i+1][j-1]
            if dp[i][j] and length > max_len:
                start = i
                max_len = length
    return s[start:start+max_len]
 

def longest_palindrome(s: str) -> str:
    #动态规划
    n = len(s)
    if n < 2:
        return s

    # 初始化二维数组 dp
    dp = [[False] * n for _ in range(n)]
    start, max_length = 0, 1

    # 所有长度为 1 的子串都是回文ß
    for i in range(n):
        dp[i][i] = True

    # 枚举子串长度
    for length in range(2, n + 1):
        for i in range(n):
            j = i + length - 1
            if j >= n:
                break
            if s[i] == s[j]:
                if length == 2:
                    dp[i][j] = True
                else:
                    #input_[i:j] <- input[i+1:j-1]
                    dp[i][j] = dp[i + 1][j - 1]

            if dp[i][j] and length > max_length:
                start = i
                max_length = length

    return s[start:start + max_length]

query = "abbadda"
result = longest_palindrome(query)
print(result)
```

```python
#给定一个二叉树，计算每一层的和，(层级遍历)
#广度优先遍历(bfs),需要借助队列数据结构对每一层进行遍历
#深度优先遍历(dfs),以递归的形式进行深度遍历
# 
# 
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def level_sums(root:Optional[TreeNode]):
    if not root:
        return []
    sums = []
    queue = [root]
    while queue:
        level_size = len(queue)
        level_sum = 0
        for _ in range(level_size):
            node = queue.pop(0)
            level_sum += node.val
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        sums.append(level_sum)
    return sums


def levelSums(root):
    if not root:
        return []

    sums = []
    queue = [root]  # 使用列表作为队列

    while queue:
        level_size = len(queue)
        level_sum = 0

        for _ in range(level_size):
            node = queue.pop(0)  # 出队
            level_sum += node.val

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        sums.append(level_sum)

    return sums

# 示例使用
# 构建一个简单的二叉树
#       1
#      / \
#     2   3
#    /|   |
#   4 5   6

root = TreeNode(1)
root.left = TreeNode(2, TreeNode(4), TreeNode(5))
root.right = TreeNode(3, right=TreeNode(6))

# 计算每层的和
print(level_sums(root)) 
```

```python
# Definition for a binary tree node.
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    #树的最大深度
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        #递归形式
        # 节点为空，高度为 0
        if root == None:
            return 0

        # 递归计算左子树的最大深度
        leftHeight = self.maxDepth(root.left)
        # 递归计算右子树的最大深度
        rightHeight = self.maxDepth(root.right)

        # 二叉树的最大深度 = 子树的最大深度 + 1（1 是根节点）
        return max(leftHeight, rightHeight) + 1

    def maxDepth_v2(self, root: Optional[TreeNode]) -> int:
        #非递归形式
        # 空树，高度为 0
        if root == None:
            return 0

        # 初始化队列和层次
        queue = [root]
        depth = 0

        # 当队列不为空
        while queue:
            # 当前层的节点数
            n = len(queue)
            # 弹出当前层的所有节点，并将所有子节点入队列
            for i in range(n):
                node = queue.pop(0)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            depth += 1
        # 二叉树最大层次即为二叉树最深深度
        return depth
```

```python
def can_measure_water(x, y, target):
    """题目描述:
    有两个水壶，容量分别为 x 和 y 升。水的供应是无限的。确定是否有可能使用这两个壶准确得到 target 升。
    你可以：
        * 装满任意一个水壶
        * 清空任意一个水壶
        * 将水从一个水壶倒入另一个水壶，直到接水壶已满，或倒水壶已空。
    1 <= x, y, target <= 10*10*10

    输入: x = 3,y = 5,target = 4
    输出: true
    """
    # 用来存放已访问的状态
    visited = set()
    # 用来进行深度优先搜索的栈，初始状态为(0, 0)
    stack = [(0, 0)]
    
    # 开始DFS
    while stack:
        a, b = stack.pop()
        
        # 如果我们已经到达了目标水量
        if a == target or b == target or a + b == target:
            return True
        
        # 如果这个状态已经访问过了，则跳过
        if (a, b) in visited:
            continue
        
        # 记录当前状态为已访问
        visited.add((a, b))
        
        # 执行各种操作并将结果推入栈中
        
        # 1. 将水壶1装满
        stack.append((x, b))
        
        # 2. 将水壶2装满
        stack.append((a, y))
        
        # 3. 清空水壶1
        stack.append((0, b))
        
        # 4. 清空水壶2
        stack.append((a, 0))
        
        # 5. 将水壶1的水倒入水壶2
        new_a = max(a - (y - b), 0)
        new_b = min(y, b + a)
        stack.append((new_a, new_b))
        
        # 6. 将水壶2的水倒入水壶1
        new_a = min(x, a + b)
        #(x-a)表示水壶1还可以倒入多少水，如果b-(x-a)小于0表示水壶2的水全部倒进了水壶1，否则就是壶2剩余的量
        new_b = max(b - (x - a), 0) 
        stack.append((new_a, new_b))
    
    # 如果遍历完所有状态都无法达到目标水量
    return False

# 测试
x = 3
y = 5
target = 8
print(can_measure_water(x, y, target))  # 输出: True
```
